Generacion de una Señal Analogica y Digital


#Primero generaremos una señal analogica y variaremos los parametros


```python
#Primero corre este bloque con shift+enter para importar librerias de python 3
#Importando librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Define los parametros abajo como ser el "tiempo_total", "A" amplitud, "f" frequencia o "theta" fase en radianes.
#Luego corre este bloque con shift+enter para generar la señal

#Parametros de la señal
A=2               #Amplitud
f=2000            #Frecuencia en Hertz
theta=1*np.pi/2   #fase en radianes

#Creando la señal

#Definiendo el tiempo
numero_de_muestras=2**16
tiempo_total=4e-3 #Segundos

vector_t=np.linspace(0,tiempo_total,int(numero_de_muestras))
Vector_Señal=A*np.sin(2*np.pi*f*vector_t+theta)

#Graficando
plt.grid(True)
plt.plot(vector_t*1000,Vector_Señal)
plt.ylabel('y(t) [v]')
plt.xlabel('t [ms]')
plt.show()
```


#Ahora generaremos una señal digital ideal de bits.

    #Puedes cambiar la taza de transmission y veras que el tamaño de los bits cambian.

    #Mayor la taza de transmision, menor la duracion del bit

    #Tambien puedes cambiar la amplitud del bit 1 en la señal

```python
#Numero de muestras
numero_de_muestras=2**16
tiempo_total=10e-9 #10 nanosegundo 


#Parametros de la señal

Amplitud_bit_1=2 #Amplitud
Tasa_de_transmission=10e9 #Tasa de transmision en bits por segundo



#Numero total de bits para el tiempo graficado
numero_de_bits=np.around(tiempo_total/(1/Tasa_de_transmission))

#Vector tiempo
vector_t=np.linspace(0,tiempo_total,int(numero_de_muestras))

#Bits aleatorios
bits=np.random.randint(2, size=int(numero_de_bits))
print("bits =" +str(bits))

#Creando la señal
Vector_Señal=Amplitud_bit_1*bits.repeat(np.ceil(numero_de_muestras/len(bits)), axis=0)

#Graficando
plt.grid(True)
plt.plot(vector_t*1e9,Vector_Señal[0:len(vector_t)])
plt.ylabel('y(t) [v]')
plt.xlabel('t [ns]')
plt.show()

```
